```python
import tweepy
import numpy as np
import pandas as pd
import json
import re
```

```python
consumer_key = "YOUR KEY HERE"
consumer_secret = "YOUR KEY HERE"

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
```

```python
access_token = "YOUR TOKEN HERE"
access_token_secret = "YOUR TOKEN HERE"

auth.set_access_token(access_token, access_token_secret)
```

```python
api = tweepy.API(auth)
```

```python
results = api.search_full_archive('ScrapeDevelopment',
                                  'socialism place:usa',
                                  fromDate = '202002010000',
                                  toDate = '202002020000',
                                  maxResults = 100)
```

```python
id_list = []
date_list = []
text_list = []

for tweet in results:
    id_list.append(tweet.id_str)
    date_list.append(tweet.created_at)
    text_list.append(tweet.text)
```

```python
Feb1SocialismDict = {'ids': id_list, 'date': date_list, 'text': text_list}
Feb1df = pd.DataFrame.from_dict(Feb1SocialismDict)
```
